```python
import pandas as pd
```

```python
data = {
    'user_id':[1,2],
    'device':[
        {'browser':'Chrome', 'os':'Windows'}, ## sturct
        {'browser':'Safari', 'os':'MacOS'},
    ],
    'interests':[
        ['sports','music'], #array
        ['movie','travel']
    ],
    'purchase':[
        [{'item':'apple','price':1.5}, {'item':'watch','price':1.0}], ## array of struct
        [{'item':'banana', 'price':1.3}]
    ]
}
```

```python
df=pd.DataFrame(data)
```

| 함수명 |기능|
|------|------|
| apply(pd.Series) | 리스트 내 딕셔너리 형태를 데이터프레임 형태로 변환 |
| explode |  리스트 내 내용을 각각의 행으로 분리 |
|

```python
## struct 컬럼을 파싱하는 방법
device_df=df['device'].apply(pd.Series) # 둘을 분리할 수 있다.
device_df # 리스트 내 딕셔너리 형태를 데이터프레임 형태로 변환
```

```python
df
```

```python
# Array 컬럼 (interest) explode

df.explode('interests') #['interests'] == 'music'
```

```python
# Array of Struct 컬럼 만들기
df_pr=df.explode('purchase')
df_pr
```

```python
pr_details=df_pr['purchase'].apply(pd.Series)
pr_details
```

```python
pd.concat([df_pr.drop(columns=['purchase']), pr_details.add_prefix('purchase_')],axis=1)
```
